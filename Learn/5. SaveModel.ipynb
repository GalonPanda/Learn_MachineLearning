```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sample_data/teams.csv')
```

```python
df
```

#### Using SkLearn Library

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
x = df[['athletes', 'age', 'prev_medals']].copy()
y = df[['medals']].copy()
```

```python
x
```

```python
y
```

```python
lr.fit(x, y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
lr.predict([[21, 22, 2]])
```

#### Save Model

```python
import pickle
```

    Dump model

```python
with open('linreg_pickle','wb') as f:
    pickle.dump(lr,f)
```

    Load dumped model

```python
with open('linreg_pickle','rb') as f:
    lp = pickle.load(f)
```

```python
lp.coef_, lp.intercept_
```
