```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris.keys()
```

```python
datas=iris.data
features=iris.feature_names
```

```python
df=pd.DataFrame(datas,columns=features)
df.sample(10)
```

```python
df['target']=iris.target
df.sample(10)
```

```python
iris.target_names
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
```

```python
df.sample(10)
```

```python
import matplotlib.pyplot as plt
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],marker='+',color='blue')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],marker='x',color='red')
plt.show()
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],marker='+',color='blue')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],marker='x',color='red')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
cls=SVC()
```

```python
x=df.iloc[:,0:4]
y=df.target
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2,random_state=2)
```

```python
cls.fit(xtrain,ytrain)
```

```python
cls.intercept_
```

```python
cls.score(xtest,ytest)
```
