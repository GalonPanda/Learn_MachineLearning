```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as lr
```

```python
data=pd.read_csv('sample_data/canada_per_capita_income.csv')
```

```python
data.head()
```

```python
x = data[['year']]
y = data[['per capita income (US$)']]
```

```python
x.head(), y.head()
```

```python
lr = lr()
```

```python
lr.fit(x, y)
```

```python
lr.coef_, lr.intercept_
```

```python
plt.title('Income Per Capita Prediction')
plt.scatter(x, y, color='blue', marker='+')
plt.plot(x, lr.predict(x), 'r-')
plt.ylabel('Income')
plt.xlabel('Year')
plt.show()
```
