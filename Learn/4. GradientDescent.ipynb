```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
total=100
x=np.array([.2*i for i in range(total)])
y=np.array([np.abs(.2*i + np.random.randn()) for i in range(total)])
```

```python
def gradesc(x,y):
    mcurr=bcurr=0
    iter=1000
    lr=.0001
    n=len(x)

    for i in range(iter):
        ypred = mcurr * x + bcurr

        cost = (1/n) * sum(val ** 2 for val in (y - ypred))
        dm = (2/n) * -1 * sum(x * (y - ypred))
        db = (2/n) * -1 * sum(y - ypred)

        mcurr = mcurr - lr * dm
        bcurr = bcurr - lr * db

        print(f'mcurr: {mcurr}, bcurr: {bcurr}, cost: {cost}, iter: {iter}')
```

```python
gradesc(x,y)
```
