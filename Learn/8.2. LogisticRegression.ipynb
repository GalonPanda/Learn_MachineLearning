```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
d = load_digits()
```

```python
d.keys()
```

```python
x = d.data
y = d.target
```

```python
plt.gray()
plt.matshow(d.images[123])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(max_iter=190)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2,random_state=99)
```

```python
log.fit(x,y)
```

```python
log.predict(xtest)
```

```python
log.score(xtest,ytest)
```

```python
from random import randint

for i in range(29):
    ran = randint(1,len(y)-1)
    yhat = log.predict([x[ran]])
    actual = y[ran]
    if actual == yhat[0]:
        print(f'Actual:Predict --> {actual}:{yhat[0]}')
        continue
    print(f'    Actual:Predict --> {actual}:{yhat[0]}')
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(log.predict(xtest),ytest)
sns.heatmap(cm,annot=True,cmap='RdYlGn')
plt.show()
```
